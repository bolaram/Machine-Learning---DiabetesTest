# Using correlation to find the same column

```python
def corr_heatmap(data_frame, size = 11):
    
    
    # Getting correlation using pandas
    correlation = data_frame.corr()
    
    #Dividing the plot into subplot for increasing the size
    fig, heatmap = plt.subplots(figsize=(size, size))
    
    #plotting the heatmap correlation heatmap
    heatmap.matshow(correlation)
    
    # Adding xticks & yticks
    plt.xticks(range(len(correlation.columns)), correlation.columns)
    plt.yticks(range(len(correlation.columns)), correlation.columns)
    
    # Displaing the graph
    plt.show()
```

# Making True to 1 and False to 0

```python
# maping the values
map_diabetes = {True: 1, False: 0}

# setting the map into the data_frames
data_frame["diabetes"] = data_frame["diabetes"].map(map_diabetes)

# showing the data again 
data_frame.head()
```

# Lets' see how many patient have their diabetes or not  

```python
num_true = len(data_frame.loc[data_frame['diabetes'] == True])
num_false = len(data_frame.loc[data_frame['diabetes'] == False])
print ("Number of patients, have Diabetes: {0} ({1:2.2f}%)".format(num_true, (num_true /(float)(num_true + num_false)) * 100))
print ("Number of patients, does not have Diabetes: {0} ({1:2.2f}%)".format(num_false, (num_false / (float)(num_true + num_false)) * 100))
```

# Splitting Data


```python
from sklearn.cross_validation import train_test_split

feature_column_names = ['num_preg', 'glucose_conc', 'diastolic_bp', 'thickness', 'insulin', 'bmi', 'diab_pred', 'age']

predicted_class_name = ['diabetes']

# Getting feature variable values

X = data_frame[feature_column_names].values
y = data_frame[predicted_class_name].values

# Saving 30% for testing
split_test_size = .30

# Splitting using scikit-learn train_test_split function

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = split_test_size, random_state = 42)
```

# let's see if the splitting works or not...

```python
print("{0:0.2f}% in training set".format((len(X_train)/float(len(data_frame.index))) * 100))
print("{0:0.2f}% in test set".format((len(X_test)/float(len(data_frame.index))) * 100))
```

# See! it's working...
# Now lets see if there any null / 0 value there...

# replace 0 with an average value using Imputer

```python
from sklearn.preprocessing import Imputer

# Impute with mean all 0 readings
fill_0 = Imputer(missing_values = 0, strategy = "mean", axis = 0)

X_train = fill_0.fit_transform(X_train) 
X_test = fill_0.fit_transform(X_test)
```

# Importing gaussian method for training

```python
from sklearn.naive_bayes import GaussianNB

# Create naive bayes model and train it with data
nb_model = GaussianNB()
nb_model.fit(X_train, y_train.ravel())
```

# Prediction from trained data

```python
# this returns array of predicted results
prediction_from_trained_data = nb_model.predict(X_train)
```

```python
# Performance metrics library
from sklearn import metrics

# Get current accuracy of the model
accuracy = metrics.accuracy_score(y_train, prediction_from_trained_data)

print("Accuracy of Naive Bayes model is: {0:.4f}%".format(accuracy * 100))
```

# Prediction from test data

```python
# this returns array of predicted results
prediction_from_test_data = nb_model.predict(X_test)

accuracy = metrics.accuracy_score(y_test, prediction_from_test_data)

print("Accuracy of Naive Bayes model is: {0:.4f}".format(accuracy))
```

This project is end here but we can do something else such as performance testing and observe the classification report of our model. And we can also try another algorithm for our dataset.

# Confusion Matrix

Usually a confusion matrix is used to describe the performance of a classification model (or "classifier") on a set of test data for which the true values are known.

```python
print("Confusion Matrix")

# lebels for set 1 = true to upper left and 0 = false to lower left
print("{0}".format(metrics.confusion_matrix(y_test, prediction_from_test_data, labels = [1, 0])))
```

see, here...

52 patients, have diabetes, are detected by our model and 52 patients actually have diabetes  <- TP

28 patients, have diabetes, are detected by our model but 28 patients really does not have diabetes  <- FP

33 patients, does not have diabetes, are detected by our model but 33 patients actually have diabetes  <- FN

118 patients, does not have diabetes, are detected by our model and 118 patients really does not have diabetes  <- TN

```python
print("Classification Report")

# labels for set 1=True to upper left and 0 = False to lower left
print("{0}".format(metrics.classification_report(y_test, prediction_from_test_data, labels = [1, 0])))
```

Precision=TP / (TP+FP)

Recall=TP / (TP+FN)

# Let's apply another algorithm for our data

```python
from numpy.core.umath_tests import inner1d
from sklearn.ensemble import RandomForestClassifier

# Create a RandomForestClassifier object
rf_model = RandomForestClassifier(random_state=42)

rf_model.fit(X_train, y_train.ravel())
```

# Accuracy Testing | Train Data

```python
rf_predict_train = rf_model.predict(X_train)

#get accuracy
rf_accuracy = metrics.accuracy_score(y_train, rf_predict_train)

#print accuracy
print("Accuracy: {0:.4f}".format(rf_accuracy))
```

# Accuracy testing for test dat

```python
rf_predict_test = rf_model.predict(X_test)

#get accuracy
rf_accuracy_testdata = metrics.accuracy_score(y_test, rf_predict_test)

#print accuracy
print ("Accuracy: {0:.4f}".format(rf_accuracy_testdata))
```

# Classification report

```python
print("Confusion Matrix for Random Forest")

# labels for set 1=True to upper left and 0 = False to lower right
print("{0}".format(metrics.confusion_matrix(y_test, rf_predict_test, labels=[1, 0])))

print("♠♠♠")

print("Classification Report\n")

# labels for set 1=True to upper left and 0 = False to lower right
print("{0}".format(metrics.classification_report(y_test, rf_predict_test, labels=[1, 0])))
```
